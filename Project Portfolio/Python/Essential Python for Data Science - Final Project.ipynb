```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
stats = df[(df["Order Date"] >= "01/01/2017") & (df["Order Date"] <= "12/31/2017")][['Sales']].agg(['sum','mean','std'])
stats
```

```python
# TODO 06 - which Segment has the highest profit in 2018
profit_2018 = df[(df["Order Date"] >="01/01/2018") & (df["Order Date"] <= "12/31/2018")].groupby(["Segment"])[["Profit"]].sum().sort_values(by = "Profit", ascending = False)
profit_2018.head(5)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
least_sales_2019 = df[(df["Order Date"] >="04/15/2019") & (df["Order Date"] <= "12/31/2019")].groupby(["State"])[["Sales"]].sum().sort_values(by = "Sales", ascending = True)
least_sales_2019.head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
## filter only 2019 and group by region 
region_sales = df[(df["Order Date"] >="01/01/2019") & (df["Order Date"] <= "12/31/2019")].groupby('Region')['Sales'].sum()
print(region_sales)

## Proportion of total sales % in west + central in 2019 
w_and_C_proportion = ((region_sales['West']+region_sales['Central'])/region_sales.sum())*100
print(round(w_and_C_proportion,2))
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
#order_2019 = df[df['Order Date'].dt.strftime("%Y") == "2019"]
#order_2020 = df[df['Order Date'].dt.strftime("%Y")  == "2020"]

# filter 2019-2020 year
filter_year = df[(df['Order Date'].dt.strftime("%Y") == "2019") | (df['Order Date'].dt.strftime("%Y")  == "2020")]

# top 10 product name in terms of total sales
top_product_sales = filter_year.groupby("Product Name")["Sales"].sum().nlargest(10).reset_index()
top_product_sales

# top 10 product name in terms of order number
top_product_order =  filter_year.groupby("Product Name")["Quantity"].sum().nlargest(10).reset_index()
print(top_product_order)
top_product_sales
```
